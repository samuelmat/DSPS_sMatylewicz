2.4 calculate a function that measures the intracluster variance (i did it in class)

```
#delete
def calICVar(X, labels):
  icvar = 0
  for n in np.unique(labels):
    icvar += np.sum((X[labels == n] - X[labels == n].mean())**2)
  return icvar
```
