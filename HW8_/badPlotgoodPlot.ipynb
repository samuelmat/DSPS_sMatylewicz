```python
import pylab as pl
import numpy as np
%pylab inline
```

```python
##second degree polynomial: arguments are i
#independent variable, dependent variable, degree of the polynomial

x = np.arange(0, 3, .02)
y1, y2 = 2 * x[::3]**2, 2 * x[::3]**2.05

c = np.exp(x)
d = c[::-1]

pl.plot(x[::3], y1, 'o', color='DarkGreen', label='one line ')
pl.plot(x[::3], y2, 'o', color='DarkBlue', label='two lines ')
pl.plot(x, c, 'k.', label='Model')
pl.plot(x, d, 'k-', label='wrong model')
pl.plot(x, c + d, 'k-', label='models sum')

#pl.plot(xnew, np.poly1d(mrnew)(xnew), 'y', alpha=1)
legend = pl.legend(loc='upper right', shadow=True, 
                   fontsize='x-large')

pl.xlabel("some IV (in fed's units)")
pl.ylabel("some DVs (in fed's units)")
```

```python
import pandas as pd
from astropy import constants as const
from astropy import units as u
```

```python
url = "https://exoplanetarchive.ipac.caltech.edu/cgi-bin/nstedAPI/nph-nstedAPI"
url +="?table=k2targets&select=epic_number,k2_dist,k2_teff,k2_rad,k2_mass,k2_bjmag,k2_vjmag,k2_gmag,k2_imag&format=CSV"

```

```python
df = pd.read_csv(url)
df.head()
```

```python
df['k2_L'] = 4 * np.pi * (df['k2_rad'] * const.R_sun)**2 * const.sigma_sb * df['k2_teff']**4 
#Adding a column for luminosity in (Watts)
```

```python
#Plot the HR diagram:
df.plot.scatter(x='k2_teff',y='k2_L', loglog=True, color='k', s=20, figsize=(9,9))
plt.xlim(40000,1000)
plt.xlabel(r'$T_{eff}$ (K)')
plt.ylabel('Luminosity (W)')
```

```python
#Plot the HR diagram:
df.plot.scatter(x='k2_teff',y='k2_L', loglog=True, color='k', s=20, alpha=0.1, figsize=(9,9))
plt.xlim(40000,1000)
plt.xlabel(r'$T_{eff}$ (K)')
plt.ylabel('Luminosity (W)')
```

```python

teff_na = df['k2_teff'].dropna()
L_na = df['k2_L'].dropna()
```
